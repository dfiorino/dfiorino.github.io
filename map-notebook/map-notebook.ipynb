```python
def GetCityDF(city):
    cityname=city.split(', ')[0]
    country=city.split(', ')[1]

    if len(city.split(', ')) > 2:
        state=country
        country=city.split(', ')[2]
        df = cities[(cities.city_ascii==cityname)&(cities.admin_name==state)&(cities.country==country)]
    else:
        df = cities[(cities.city_ascii==cityname)&(cities.country==country)]
    return df.iloc[0]
```

```python
m = folium.Map(
    location=[14.511940, -38.945430],
    zoom_start=2
)

minimap = foliumplugins.MiniMap()
m.add_child(minimap)

feature_group = folium.FeatureGroup(control=False)
m.add_child(feature_group)



for key_type,val_places in places_dict.items():
    subgroup = folium.plugins.FeatureGroupSubGroup(feature_group, key_type)
    m.add_child(subgroup)
    for place,sub_places in val_places[0].items():
        citydf = GetCityDF(place)

        popup_html = folium.Html('<strong>{}:</strong><br>'.format(place)+'<br>    '.join(sub_places),
                                  script=True)
        popup = folium.Popup(popup_html,max_width=2650,
                            parse_html=True)

        folium.CircleMarker([citydf.lat, citydf.lng],
                            radius = 8,
                            fill=True,
                            fill_color=val_places[1],
                            color = val_places[1],
                            fill_opacity=.85,
                            popup=popup).add_to(subgroup)     


l = folium.LayerControl().add_to(m)

m.save('travels.html')
m
```
