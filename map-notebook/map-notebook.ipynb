```python
def GetCityDF(city):
    cityname=city.split(', ')[0]
    country=city.split(', ')[1]

    if len(city.split(', ')) > 2:
        state=country
        country=city.split(', ')[2]
        df = cities[(cities.city_ascii==cityname)&(cities.admin_name==state)&(cities.country==country)]
    else:
        df = cities[(cities.city_ascii==cityname)&(cities.country==country)]
    return df.iloc[0]
```

```python
m = folium.Map(
    location=[0,0],
    zoom_start=2
)

minimap = foliumplugins.MiniMap()
m.add_child(minimap)


mcg = folium.plugins.MarkerCluster(control=False)
m.add_child(mcg)


for k,v in places_dict.items():
    subgroup = folium.plugins.FeatureGroupSubGroup(mcg, k)
    m.add_child(subgroup)
    
    for v_v in v[0]:
        citydf = GetCityDF(v_v)
#         print(citydf)
        folium.CircleMarker([citydf.lat, citydf.lng],
                                    radius = 6,
                                    fill=True,
                                    fill_color=v[1],
                                    color = v[1],
                                    fill_opacity=0.7).add_to(subgroup)

        

l = folium.LayerControl().add_to(m)

m.save('where-in-the-world.html')
```
